```python
from queue import Queue
#Codigo BFS:
def BFS(lista_adj, raiz):
    #Criaremos dicionarios para os vertices visitados, seu nivel, lista de pais
    visitados = {}
    nivel = {}
    pai = {} #nos ajudara a encontrar caminho e distancia minima
    bfs_output=[]
    queue = Queue()

    for vertice in lista_adj.keys():
        #inicializando:
        visitados[vertice] = False
        nivel[vertice] = -1 #equivalente a ideia de usar infinito
        pai[vertice] = None
    
    #Condiçao da raiz:
    s = raiz
    visitados[s] = True
    nivel[s] = 0
    queue.put(s)
    
    #exploraçao:
    while not queue.empty():
        u = queue.get()
        bfs_output.append(u)
        
        #analisando os vizinhos de u:
        for v in lista_adj[u]:
            if not visitados[v]: #verificaçao se v ja foi visitado
                visitados[v] = True
                pai[v] = u
                nivel[v] = nivel[u] + 1
                queue.put(v)
    
    return bfs_output, nivel, pai
```

## 3) Escrevendo os pesos das arestas em uma matriz esparsa

Para isso criaremos um dicionário

Se as arestas do grafo não tiverem pesos:

$\texttt{retorna None}$
    
Se as arestas do grafo tiverem pesos:

$\texttt{retorna um dicionario com chave = aresta e item da chave igual o peso (em float)}$

```python
def arestas_pesos(arestas):
    mat = {} #definindo o dicionario
    W = {}
    if len(arestas[0]) == 2:
        return None
    lista = []
    for u,v,w in arestas:
        if ((int(v),int(u)) not in mat) and (v != u):
            mat[(int(u),int(v))] = float(w) #Usaremos as chaves como tuplas, já que dicionario não aceita lista como chave
            lista.append((int(u),int(v)))
            
    lista.sort()
    
    for E in lista:
        u,v = E
        W[(str(u),str(v))] = mat[u,v]
    return W
```

### 4.1) Dijkstra

Primeiro vamos escrever a função de Dijkstra:

$\texttt{Algoritimo de Dijkstra}$

$Entrada:$ lista com vertices, matriz com pesos, vertice origem(s) 

$Saída:$ vetor distancia dos vertices até s e vetor de pais

```python
#Função de Dijkstra:
def Dijkstra(lista_adjacencia_pesos, s):
    n = len(lista_adjacencia_pesos.keys())
    s = int(s)
    inf = float('inf')
    list_adj = lista_adjacencia_pesos.copy()
    for V in list_adj.keys(): #custo de ir de V ate V é zero
        list_adj[V][V] = 0
    
       
    
    ##### Iniciando codigo
    marcados = n*[0] #nenhum vertice marcado
    dist = n*[inf] #Vetor que armazena as distancias. Trataremos simbolicamente infinito como -1
    dist[s-1] = 0 # distancia da raiz até ela mesma é zero
    D = [(0,s)] #Guardando a posição do heap da forma (dist(w), w), ou seja, a distancia de s a w é dist(w)
    
    for v in range(1,n+1):# Colocando no heap que a distancia de s até os outros vertices é inf
        if v != s: (heapq.heappush(D, (dist[v-1], v)))
            
    pai = n*[None]
    lista = []
    #print(D)                       
    ##### Loop:
    while D != []:
        dist_min, u  = heapq.heappop(D) # retiro vertice de distancia minima
        marcados[u-1] = 1 # marcando u
        
        for v in list_adj[str(u)]:
            v = int(v)
            #print(v)
            if marcados[v-1] == 0:
                if dist[u-1] + list_adj[str(u)][str(v)] < dist[v-1]:
                    ### Achando posiçao de v no heap
                    for i in range(len(D)):
                        if D[i] == (dist[v-1],v):
                            pos = i
                            break
                    
                    ### Atualizando:
                    
                    dist[v-1] = dist[u-1] + list_adj[str(u)][str(v)]
                    D[pos] = (dist[v-1],v)
                    heapq._siftdown(D, 0, pos) #atualiza heap
                    pai[v-1] = str(u)

    return dist, pai
    
    
```

```python
def dist_caminho_s(lista_adjacencia, lista_adj_peso, W, s):
    resposta = {}
    for V in lista_adjacencia.keys():
        path, dist_V = distancia_par(lista_adjacencia, lista_adj_peso, W, s, V)
        d = '({one},{two})'.format(one = str(s), two = V)
        resposta[d] = {'distancia': dist_V, 'Caminho': path}
    return resposta
```

## 5) Árvore Geradora Mínima

Para resolver esse problema aplicaremos o $\texttt{Algoritmo de Kruskal}$

$\texttt{Entrada: }$ Número de vértices e Dicionário com as arestas com respectivos pesos

$\texttt{Saída: }$ Arestas da MST, MST e custo da MST

```python
def MST(numero_vertices, W):
    n = numero_vertices
    
    inicio = time.time()
    ##### Arestas da MST
    arestas_MST = []
    
    ##### Iniciando heap:
    H = []
    for E in W.keys():
        u = int(E[0])
        v = int(E[1])
        w = W[E]
        heapq.heappush(H, (w, u, v)) ## colocando aresta no heap
    
    ##### Criando conjunto C:
    C = [[]*n for i in range(1,n+1)]
    for i in range(1,n+1):
        C[i-1].append(i) #Alocando de forma que cada conjunto comece com um vertice
    
    ##### Para saber onde os vertices estarão nos conjuntos ao longo da implementação, precisamos de um hash:
    S = [] #S[i] é o conjunto ao qual o vertice i pertence
    for i in range(1,n+1):
        S.append(i)
        
    #print(C)
    #print(S)
        
    ##### Parametros:
    cont_E = 0 ## Contador de arestas
    custo = 0
    
    ##### Iniciando o loop:
    while cont_E < n-1:
        w, u, v = heapq.heappop(H)
        if S[u - 1] != S[v - 1]: ## Se conjunto(ou arvore) em que está o vertice u é diferente da arvore que está o vertice v
            custo = custo + w
            r = S[u - 1]
            s = S[v - 1]
            if s <= r:
                r, s = s, r
            for j in C[s - 1]:
                S[j - 1] = r
            
            ### Unindo C[r] e C[s]
            C[r - 1].extend(C[s-1])
            C[s - 1] = []
            cont_E += 1
            arestas_MST.append((str(u),str(v)))
            
        ### Visualizando como a arvore é montada
        #print(C)
        #print(S)
    
    final = time.time()
    tempo = final - inicio
    print('tempo: {}'.format(str(tempo)))
    return arestas_MST, C[0], custo
    
    
    
    
```

```python
def imprimir_MST(grafo_W, numero_vertices, W):
    arestas_MST, _, custo = MST(numero_vertices, W)
    l = [custo]
    for E in arestas_MST:
        l.append(E)
        
    with open('MST_{}'.format(str(grafo_W)), 'w') as arquivo:
        for u in l:
            arquivo.write(str(u) + '\n')
    
    print(arquivo)
    return arquivo
```
